```python

# Import necessary libraries
import pandas as pd

# Load the dataset from your Google Drive or Colab environment
# Replace the file path below with your provided path
df = pd.read_csv('/content/drive/MyDrive/Data Analytics Projects/Data Cleaning Project/Data/raw_data.csv', encoding='ISO-8859-1')

# 1. Remove rows with missing descriptions
df_cleaned = df.dropna(subset=['Description'])

# 2. Remove rows with missing customer IDs
df_cleaned = df_cleaned.dropna(subset=['CustomerID'])

# 3. Remove rows with negative quantities
df_cleaned = df_cleaned[df_cleaned['Quantity'] >= 0]

# 4. Remove rows with zero or negative unit prices
df_cleaned = df_cleaned[df_cleaned['UnitPrice'] > 0]

# 5. Remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates()

# Save the cleaned data to a new CSV file with the corrected path
df_cleaned.to_csv('/content/drive/MyDrive/Data Analytics Projects/Data Cleaning Project/Data/final_cleaned_data.csv', index=False)

# Output file path for reference
print("Cleaned data saved to: /content/drive/MyDrive/Data Analytics Projects/Data Cleaning Project/Data/final_cleaned_data.csv")
```
